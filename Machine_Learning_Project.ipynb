```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv('USA_cars_datasets.csv')
df.dropna()
df.brand.unique()
df.dtypes
df['Average Price']=df.groupby('brand')['price'].transform('mean')
df

```

```python
df.brand.unique()
df.dtypes
```

```python
sns.barplot(data=df,x='Average Price',y='brand')
```

```python
sns.lineplot(data=df,x='mileage',y='price')
```

```python
sns.countplot(y='brand', data = df)
```

```python
g = sns.FacetGrid(df, col="title_status", hue="title_status")
g.map(sns.regplot, "price", "year").add_legend()
plt.show()
```
